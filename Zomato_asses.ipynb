```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset
df=pd.read_csv("zomato_rest_data.xlsx - Sheet1.csv")
df
```

#Task-1 Data Understanding & Cleaning

**1)Dataset Exploration**

```python
#Display first 5 rows
df.iloc[:5,:]
```

```python
#Display shape of dataset (rows and columns)
df.shape
```

```python
#List column names
df.columns
```

```python
#Check data types
df.info()
```

```python
#Identify missing values
df.isnull().sum()
```

**2. Data Cleaning**

```python
#dropping duplicates
df.drop_duplicates(inplace=True)
df.shape
```

```python
#drop unnecessary column
df=df.drop("Unnamed: 0",axis=1)
df=df.drop("url",axis=1)
df=df.drop("address",axis=1)
df=df.drop("phone",axis=1)
df=df.drop("location",axis=1)
df=df.drop("menu_item",axis=1)
df=df.drop("reviews_list",axis=1)
df=df.drop("rest_type",axis=1)
```

```python
#handle rate column remove /5 and handle new and - and change to float and handling null values
#1)removing /5
df["rate"]=df["rate"].str.replace("/5","")
#2)handling new and -
df["rate"]=df["rate"].replace("-",np.nan)
df["rate"]=df["rate"].replace("NEW",np.nan)
#3)changing data type
df["rate"]=df["rate"].astype(float)
#4)handling null values
mean_rate = round(df['rate'].mean())
df['rate'] = df['rate'].fillna(mean_rate)
```

```python
#changing datatype of approxcost and handling null values
# 1)removing comma
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].str.replace(",","")
#2)changing datatype
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(float)
#3)handling null values
mode_approx_cost =df['approx_cost(for two people)'].mean()
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(mode_approx_cost)
```

```python
#rename listed_in(city) column with location
df=df.rename(columns={"listed_in(city)":"location"})
```

```python
#rename listed_in(type) column with rest_type
df=df.rename(columns={"listed_in(type)":"rest_type"})
```

```python
#handling null values in dishliked with mode
mode_dishliked = df['dish_liked'].mode()[0]
df['dish_liked'] = df['dish_liked'].fillna(mode_dishliked)
```

```python
#handling null values in Cuisine with mode
mode_cuisine = df['cuisines'].mode()[0]
df['cuisines'] = df['cuisines'].fillna(mode_cuisine)
```

```python
# resetting the index
df.reset_index(drop=True,inplace=True)
```

#Task 2: Univariate Analysis

```python
# 1. Top 5 locations with highest number of restaurants and hist plot
top5=df["location"].value_counts()
plt.bar(top5.index,top5.values,color="red")
plt.xlabel("Location")
plt.ylabel("Number of Restaurants")
plt.title("Top 5 Locations with Highest Number of Restaurants")
plt.xticks(rotation=90)
plt.show()
```

**Interpretation**
- Location BTM contains more number of Restaurants.

```python
#Distribution of restaurant ratings either valuecounts or histplot

sns.histplot(df["rate"],color="red")
plt.xlabel("Ratings")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Restaurant Ratings")
plt.show()
```

**Interpretation**
- The graph shows the normal distribution
- 12000 plus restaurants are having max rating of 4.0



```python
#Percentage of restaurants offering online orders with pie chart
order_count=df["online_order"].value_counts()
plt.pie(order_count,labels=order_count.index,autopct="%0.2f%%",colors=["red","green"])
plt.show()
```

**Interpretation**
- 58.87% of users are prefering online orders

```python
#4. Distribution of average cost for two
avg_cost=df.groupby("location")["approx_cost(for two people)"].mean().sort_values(ascending=False)
sns.histplot(avg_cost,color="r")
plt.xlabel("Cost at that Location")
plt.ylabel("Average Cost")
plt.title("Distribution of Average Cost for Two")
plt.xticks(rotation=90)
plt.show()
```

**Interpretation**
- The graph shows the normal distribution
- 12 locations have avg cost in the range of 460 and 520

# Task 3: Bivariate Analysis

```python
#Rating vs Cost
plt.scatter(x="rate",y="approx_cost(for two people)",data=df,color="r")
plt.xlabel("Rating")
plt.ylabel("Cost")
plt.title("Rating vs Cost")
plt.show()
```

**Interpretation**
- Average cost upto 2000 having more no of ratings.
- If average cost is more rating is more

```python
# Rating vs Online Order
order=df.groupby("online_order")["rate"].mean()
plt.bar(order.index,order.values,color="r")
plt.title('Rating vs Online Order')
plt.xlabel('Online Order Available')
plt.ylabel('Rate')
plt.show()
```

**Interpretation**
- Customers are giving equal rating for online order or not

```python
#Votes vs Rating
plt.scatter(x="rate",y="votes",data=df,color="r")
plt.xlabel("Rating")
plt.ylabel("Votes")
plt.title("Rating vs Votes")
plt.show()
```

**Interpretation**
- More no of votes are there for highest ratings.
- 5000 plus votes are there between 4 to 5 rating
